```python
# Problem 1
import random
import matplotlib.pyplot as plt
x = 0
step = [i for i in range(1000)]
pos = []
for i in range(1000):
    if x == -4:
        x = x + 1
        pos.append(x)
    else:
        if random.random() < 0.5:
            x = x + 1
        else:
            x = x - 1
        pos.append(x)
plt.subplot(1,2,1)
plt.scatter(step, pos)
plt.subplot(1,2,2)
plt.hist(pos)
plt.show()
```

```python
# Problem 2
import copy

import math
import numpy as np
def brownian(steps, var):
    x = [0, 0] # A()
    step = [i for i in range(steps)]
    pos = []
    for i in range(steps):
        direction = np.random.randint(0, 360) * math.pi / 180
        length = np.random.normal(0, math.sqrt(var))
#         print(direction, length)
        x[0] = x[0] + length*math.cos(direction)
        x[1] = x[1] + length*math.sin(direction)
        
        z = copy.deepcopy(x)
#         print(x)
        pos.append(z)
#         print(pos)
    x_coor = [each[0] for each in pos]
    y_coor = [each[1] for each in pos]
    return pos[-1]
#     plt.scatter(x_coor, y_coor)
#     plt.show
#     print(pos)
brownian(1600, 1)
```
