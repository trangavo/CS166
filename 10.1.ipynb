```python
# Gamblers
import numpy as np
def Gambler(stop_time):
    playing_time = []
    for j in range(20):
        R = [100]
        i = 0
        steps = [0]
        while (R[i] > 0 and R[i] < 2100 and i < stop_time):
            i = i+1
            steps.append(i)
            w = 1 if random.random() < 0.5 else -1
            R.append(R[i-1] + w)
        playing_time.append(len(R)/(10**5))
    plt.hist(playing_time)
    playing_time = np.array(playing_time)
    mean = np.mean(playing_time) * (10**5)
    variance = np.var(playing_time) * (10**5)
    print(mean, variance)
plt.subplot(2,2,1)
Gambler(1000000)
plt.subplot(2,2,2)
Gambler(2000000)
plt.subplot(2,2,3)
Gambler(3000000)
plt.subplot(2,2,4)
Gambler(4000000)
```
